## 11 Change data type

Data Type in Python is as it follows:
    
><i>For character variable:<b><font color=blue>object</font></b><br>
>For categorical variable: <b><font color=blue>category</font></b><br>

```python
# Using a "function".
```

```python
def coerce_df_columns_to_numeric(df, column_list):
    df[column_list] = df[column_list].apply(pd.to_numeric, errors='coerce')
```

This function allows more than one column at a time.

```python
df2 = df # To keep df original stucture
```

```python
coerce_df_columns_to_numeric(df2, ['col2'])
```

```python
df2.dtypes
```

```python
df2
```

```python
# Example 2: Various fields to numeric
```

```python
df2 = df # To keep df original stucture
```

```python
coerce_df_columns_to_numeric(df2, ['col2', 'cost','s'])
```

```python
df2.dtypes
```

```python
df2
```

```python
Note that if the field has some character data then the data is coerce as invalid values to NaN.
```
