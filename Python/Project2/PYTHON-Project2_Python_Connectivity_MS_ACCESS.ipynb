## 3 How about Creating a Database?

```python
import pandas as pd
from os import getcwd
import pyodbc

# Complete path of the database
DB_PATH = getcwd() + "\Base\Innova_reduce.mdb"
db_file = r'DB_PATH'
user = 'user'
password = 'pw'

# Connect to the database
conn = pyodbc.connect('DRIVER={{Microsoft Access Driver (*.mdb, *.accdb)}};DBQ=' + \
                      '{};Uid={};Pwd={};'.format(db_file, user, password)

# Select elements new table

sql = "SELECT rol, region, ventas_ao_2005, empleo_ao_2005 FROM innova_first"

# Made pd data frame ("paso")

paso = pd.read_sql_query(sql, conn)

# Create new table "first333" in "Innova_reduce" database 

paso.to_sql(name='first333', con=conn, if_exists='replace',index=False)

# Verify if new table "first333" was created in "Innova_reduce" database

sql = "SELECT * FROM first333"
df = pd.read_sql_query(sql, conn)
print (df.shape)
df.head()
conn.close()
```

## 4 How about Quering by JOIN and Aggregations Summary?

### 41 Joining tables with JOIN

```python
### Simple JOIN

import pandas as pd
from os import getcwd
import pyodbc

# Complete path of the database
DB_PATH = getcwd() + "\Base\Innova_reduce.mdb"
db_file = r'DB_PATH'
user = 'user'
password = 'pw'

# Connect to the database
conn = pyodbc.connect('DRIVER={{Microsoft Access Driver (*.mdb, *.accdb)}};DBQ=' + \
                      '{};Uid={};Pwd={};'.format(db_file, user, password)

sql = "SELECT innova_first.rol, innova_first.region, tamano, ventas_ao_2005 from innova_first inner join innova_second on innova_first.rol = innova_second.rol"
df = pd.read_sql_query(sql, conn)
print (df.shape)
df.head()
conn.close()
```

```python
### JOIN with agregated summary

import pandas as pd
from os import getcwd
import pyodbc

# Complete path of the database
DB_PATH = getcwd() + "\Base\Innova_reduce.mdb"
db_file = r'DB_PATH'
user = 'user'
password = 'pw'

# Connect to the database
conn = pyodbc.connect('DRIVER={{Microsoft Access Driver (*.mdb, *.accdb)}};DBQ=' + \
                      '{};Uid={};Pwd={};'.format(db_file, user, password)

sql = "SELECT innova_first.region, tamano, AVG(ventas_ao_2005) AS AVERAGE, stdev(ventas_ao_2005) as DEstandar, COUNT(ventas_ao_2005) as MUESTRA,  MAX(ventas_ao_2005) AS MAXIMO, MIN(ventas_ao_2005) AS MINIMO from innova_first  inner join innova_second on innova_first.rol = innova_second.rol GROUP BY  innova_first.region, tamano  order BY innova_first.region, tamano"
df = pd.read_sql_query(sql, conn) 
print (df.shape)
df.head()
conn.close()
```
